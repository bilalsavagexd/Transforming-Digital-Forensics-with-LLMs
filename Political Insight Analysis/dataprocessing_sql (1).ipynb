```python
from google.colab import files
import sqlite3

# Step 1: Upload the SQLite database file
uploaded = files.upload()

# Step 2: Get the uploaded filename
filename = list(uploaded.keys())[0]

# Step 3: Connect to the uploaded SQLite database
conn = sqlite3.connect(filename)
cursor = conn.cursor()

# Step 4: Query to fetch rows containing the keyword "Israel" in the "ExtractedBodyText" column
query = "SELECT ExtractedBodyText FROM emails WHERE lower(ExtractedBodyText) LIKE '%israel%';"
cursor.execute(query)

# Fetch all rows
rows = cursor.fetchall()

# Define the file name
file_name = "email_israel.txt"

# Step 5: Write the results to a text file
with open(file_name, "w") as file:
    for row in rows:
        file.write(str(row) + "\n")

# Step 6: Close the connection
conn.close()

# Provide the file to download (useful in Colab)
files.download(file_name)
```
